```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Worldometers_data.csv")
```

```python
nw_data = data.loc[:, data.columns != 'Country']
```

```python
def columnrow(col):
        return col.str.replace(r'[^\d.]+', '').fillna(0).astype(float)
```

```python
df = nw_data.apply(columnrow, axis=0)
```
